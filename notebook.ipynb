```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dateutil import parser

%matplotlib inline
```

```python
#read all extracted data in utf8 format
data = pd.read_csv('extracted_data.csv')
print(data.shape)
data.head(10)
```

```python
#pie chart of rating counts
rating_counts = data['Rating'].value_counts()

# Plot the pie chart
plt.figure(figsize=(10, 7))
plt.pie(rating_counts, labels=None, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Pie Chart of all Article Ratings on FactsFirst')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.legend(labels=rating_counts.index, title = "Article Ratings")
plt.show()
```

```python
#bar chart of rating counts
rating_counts = data['Rating'].value_counts()

ax = rating_counts.plot(kind='bar', figsize=(10,7))
```

```python
# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Extract year and month from the 'Date' column
data['YearMonth'] = data['Date'].dt.to_period('M')

# Count the number of articles per month
monthly_counts = data['YearMonth'].value_counts().sort_index()

# Plot the line graph
plt.figure(figsize=(12, 6))
plt.plot(monthly_counts.index.astype(str), monthly_counts.values, marker='o', linestyle='-')
plt.title('Number of Uploaded Articles on FactsFirst Each Month')
plt.xlabel('Month')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Display the line graph
plt.show() 
```
